# Python | Visualize missing values (NaN) values using Missingno Library
Last Updated : 04 Jul, 2019

現実世界のデータセットの場合、データセット内の一部の値が欠落していることがよくあります。 これらの欠損値を NaN (Not a Number) 値として表します。 ただし、優れた機械学習モデルを構築するには、データセットが完全である必要があります。 そのため、いくつかの代入手法を使用して、NaN 値をいくつかの可能性のある値に置き換えます。 ただし、その前に、NaN 値がデータセット内でどのように分布しているかをよく理解する必要があります。

**Missingno** ライブラリは、NaN 値の分布を視覚化する非常に優れた方法を提供します。 Missingno は Python ライブラリであり、Pandas と互換性があります。

**Install the library –**
```py
pip install missingno
```
コードで使用されるデータセットを取得するには、[ここ](https://openmv.net/info/kamyr-digester) をクリックしてください。

## Matrix :
この行列を使用すると、データセット内の欠損パターンを非常に迅速に見つけることができます。 この例では、列 AAWhiteSt-4 と SulphidityL-4 には同様の欠損値パターンがありますが、UCZAA は異なるパターンを示しています。

````python
# Program to visualize missing values in dataset 
  
# Importing the libraries 
import pandas as pd 
import missingno as msno 
  
# Loading the dataset 
df = pd.read_csv("./data/kamyr-digester.csv") 
  
# Visualize missing values as a matrix 
msno.matrix(df) 
````

## Bar Chart :
この棒グラフは、各列に欠損値がいくつあるかを示します。 この例では、AAWhiteSt-4 と SulphidityL-4 に欠損値が最も多く含まれており、次に UCZAA が続きます。

````python
# Program to visualize missing values in dataset 
  
# Importing the libraries 
import pandas as pd 
import missingno as msno 
  
# Loading the dataset 
df = pd.read_csv("./data/kamyr-digester.csv") 
  
# Visualize the number of missing 
# values as a bar chart 
msno.bar(df) 
````

## Heatmap :
ヒートマップは、2 列ごとの欠損の相関関係を示します。 この例では、AAWhiteSt-4 と SulphidityL-4 の間の相関関係は 1 です。これは、それらの一方が存在する場合、もう一方も存在する必要があることを意味します。

* **-1** に近い値は、1 つの変数が出現する場合、もう 1 つの変数が **欠落している可能性が非常に高いことを意味します**。
* **0** に近い値は、2 つの変数の欠損値の出現の間に **依存性がない**ことを意味します。
* **1** に近い値は、1 つの変数が出現する場合、もう一方の変数も **存在する可能性が非常に高いことを意味します**。

````python
# Program to visualize missing values in dataset 
  
# Importing the libraries 
import pandas as pd 
import missingno as msno 
  
# Loading the dataset 
df = pd.read_csv("./data/kamyr-digester.csv") 
  
  
# Visualize the correlation between the number of 
# missing values in different columns as a heatmap 
msno.heatmap(df) 
````
